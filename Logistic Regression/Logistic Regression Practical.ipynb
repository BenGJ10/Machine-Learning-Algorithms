```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
# Making Datasets
X, Y = make_classification(n_samples = 1000, n_features = 10, n_classes = 2, random_state = 15)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state = 42)
```

### Model Training

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(X_train, Y_train)
```

```python
Y_pred = logistic.predict(X_test)
```

```python
print(Y_pred)
```

### Performance Metrics

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
score = accuracy_score(Y_pred, Y_test)
print(score)
print(classification_report(Y_pred, Y_test))
print(confusion_matrix(Y_pred, Y_test))
```

### Hyperparameter Tuning and Cross Validation

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
parameters = dict(penalty = penalty, C = c_values, solver = solver)
```

<b>Grid Search CV</b>

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator = model, param_grid = parameters, scoring = "accuracy", cv = cv, n_jobs = -1)
```

```python
grid
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
Y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(Y_pred, Y_test)
print(score)
print(classification_report(Y_pred, Y_test))
print(confusion_matrix(Y_pred, Y_test))
```

We can see that the score increased to 0.92.

<b> Randomized Search CV </b>

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator = model, param_distributions = parameters, cv=5, scoring='accuracy')
```

```python
randomcv.fit(X_train, Y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
Y_pred = randomcv.predict(X_test)
```

```python
score = accuracy_score(Y_pred, Y_test)
print(score)
print(classification_report(Y_pred, Y_test))
print(confusion_matrix(Y_pred, Y_test))
```
